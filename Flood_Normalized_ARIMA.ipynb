Normalize River factor Data

```python
for i in range(0,1080):
  df['Rain'][i]=(df['Rain'][i]-df['Rain'].min())/(df['Rain'].max()-df['Rain'].min())
```

Normalize Temperature Data

```python
for i in range(0,1080):
  df['Temp'][i]=(df['Temp'][i]-df['Temp'].min())/(df['Temp'].max()-df['Temp'].min())
```

Normalize Assam rain data

```python
# normalized_df=(df-df.min())/(df.max()-df.min())
```

```python
for i in range(0,1080):
  df['Rain Assam'][i]=(df['Rain Assam'][i]-df['Rain Assam'].min())/(df['Rain Assam'].max()-df['Rain Assam'].min())
```

```python
df
```

Normalise Population

```python
for i in range(0,1080):
  df['Population'][i]=(df['Population'][i]-df['Population'].mean())/(df['Population'].std())
```

```python
import pmdarima as pm
```

```python
model = pm.auto_arima(df['Flood'], 
                        m=12, seasonal=True,test='adf',error_action='ignore',  
                           suppress_warnings=True, trace=True)
```

```python
print(df.shape)
train=df.iloc[:-380]
test=df.iloc[-380:]
print(train.shape,test.shape)
```

```python
test
```

```python
model.fit(train['Flood'])
```

```python
forecast=model.predict(n_periods=380, return_conf_int=True)
```

```python
forecast
```

```python
forecast_df = pd.DataFrame(forecast[0],index = test.index,columns=['Prediction'])
```

```python
forecast_df
```

```python
test.iloc[:, 1]
```

```python
error = 0
for i in range(0,380):
  if(forecast_df.iat[i,0]<0.5):
    forecast_df.iat[i,0]=0
  else:
    forecast_df.iat[i,0]=1
```

```python
pd.concat([test['Flood'],forecast_df],axis=1).plot()
```

```python
error = 0
for i in range(0,300):
  if(test['Flood'][i]-forecast_df.iat[i,0]>0):
    error=error+1
```

```python
accuracy = 1-error/test.shape[0]
```
